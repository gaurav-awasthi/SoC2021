```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('/content/Graduate Admissions.csv')
df
```

```python
df.columns
```

Influence of GRE and TOEFL scores on chance of getting admitted

```python
sns.lmplot(data=df,x='GRE Score',y='Chance of Admit ')
plt.show()
sns.lmplot(data=df,x='TOEFL Score',y='Chance of Admit ')
plt.show()
```

Influence of GRE and TOEFL scores on research experience

```python
sns.boxplot(x='Research',y='GRE Score',data=df)
plt.show()
sns.boxplot(x='Research',y='TOEFL Score',data=df)
plt.show()
```

Influence of university rating on SOP and LOR rating

```python
sns.violinplot(data=df,x='University Rating',y='SOP')
plt.show()
sns.violinplot(data=df,x='University Rating',y='LOR ')
plt.show()
```

Average SOP and LOR rating of admitted students, assuming admitted students have a >=0.95 chance of admit

```python
df_admits = df[df['Chance of Admit ']>=0.95]
sns.distplot(df_admits['GRE Score'])
plt.show()
print('The average GRE score is',df_admits['GRE Score'].mean())
sns.distplot(df_admits['TOEFL Score'])
plt.show()
print('The average TOEFL score is',df_admits['TOEFL Score'].mean())
```

Influence of university rating on chance of getting admitted

```python
sns.boxplot(data=df,x='University Rating',y='Chance of Admit ')
plt.show()
```

GRE and TOEFL scores for a 90% chance of admit

```python
gre = []
toefl = []
for r in df.index:
  if (df.loc[r,'Chance of Admit '] == 0.9):
    gre.append(df.loc[r,'GRE Score'])
    toefl.append(df.loc[r,'TOEFL Score'])

sns.distplot(gre)
plt.show()
sns.distplot(toefl)
plt.show()
print('The mean GRE score is:',sum(gre)/len(gre))
print('The mean TOEFL score is:',sum(toefl)/len(toefl))
```
